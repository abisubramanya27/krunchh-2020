```python
print(dfMain['Sales'].dtype)
dfMain['Sales'] = pd.to_numeric(dfMain['Sales'].str.replace(',',''), errors='raise')
dfMain.isnull().sum()
```

```python
df = dfMain.copy()[['Order Date','Category','Discount', 'Number of Records', 'Profit','Quantity', 'Sales', 'Shipping Cost']]
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.sort_values(by = 'Order Date',ascending = True,inplace = True)
df.head(20)
```

```python
from pylab import rcParams
import matplotlib.pyplot as plt

rcParams['figure.figsize'] = 16, 8
rcParams['axes.labelsize'] = 14
rcParams['xtick.labelsize'] = 12
rcParams['ytick.labelsize'] = 12
rcParams['text.color'] = 'k'
```

```python
furniture = df[df['Category'] == 'Furniture'][['Order Date','Sales']].groupby('Order Date').sum()
office = df[df['Category'] == 'Office Supplies'][['Order Date','Sales']].groupby('Order Date').sum()
tech = df[df['Category'] == 'Technology'][['Order Date','Sales']].groupby('Order Date').sum()
tech.head()
```

```python
furniture['Sales'].rolling(15).mean().plot()
```

# Seasonal Decompose

```python
yFurn = furniture['Sales'].resample('MS').mean()
yOffice = office['Sales'].resample('MS').mean()
yTech = tech['Sales'].resample('MS').mean()

yOffice.plot()
```

```python
yFurn.plot()
```

```python
import statsmodels.api as sm

decompFurn = sm.tsa.seasonal_decompose(yFurn, model='additive')
figFurn = decompFurn.plot()
plt.show()
```

```python
yOffice.plot()
```

```python
decompOffice = sm.tsa.seasonal_decompose(yOffice, model='additive')
figOffice = decompOffice.plot()
plt.show()
```

```python
yTech.plot()
```

```python
decompTech = sm.tsa.seasonal_decompose(yTech, model='additive')
figTech = decompTech.plot()
plt.show()
```

# Furniture sales fit

```python
import itertools

p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(yFurn,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                           )
            results = mod.fit()
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

### ARIMA(0,1,1) (0,1,1,12) yields lowest AIC of 541.77

```python
tmpFurn = yFurn.diff(12)[12:].diff()[1:]
tmpFurn.plot()
```

# ACF

```python
acf_val = sm.tsa.stattools.acf(tmpFurn)
num_lags = 20
plt.bar(range(num_lags),acf_val[:num_lags])
plt.axhline(y = 0.2,linestyle = '--')
plt.axhline(y = -0.2,linestyle = '--')
plt.axhline(y = 0,linestyle = '-')
```

# PCF

```python
pacf_val = sm.tsa.stattools.pacf(tmpFurn,nlags = 16)
num_lags = 16
plt.bar(range(num_lags),pacf_val[:num_lags])
plt.axhline(y = 0.2,linestyle = '--')
plt.axhline(y = -0.2,linestyle = '--')
plt.axhline(y = 0,linestyle = '-')
```

```python
mod = sm.tsa.statespace.SARIMAX(yFurn,
                        order=(3,2,1),
                        seasonal_order=(0,1,1,12),
                        )
result = mod.fit()
print(result.aic)
```

```python
from datetime import datetime
from datetime import timedelta

Furn_my_order = (0,1,1)
Furn_my_seasonal_order = (0,1,1,12)

# Furn_train = yFurn[:datetime(2014,9,1)]
# Furn_test = yFurn[datetime(2014,9,1) + timedelta(days = 1):]
Furn_mod = sm.tsa.statespace.SARIMAX(yFurn,
                                order=Furn_my_order,
                                seasonal_order=Furn_my_seasonal_order,
                                )
Furn_res = Furn_mod.fit()
print(Furn_res.summary())
```

```python
Furn_res.plot_diagnostics(figsize=(16, 8))
plt.show()
```

# Furniture Sales Forecast for next 2 years

```python
pred = Furn_res.get_prediction(start=datetime(2014,1,1), dynamic=False)
pred_ci = pred.conf_int()
ax = yFurn.plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```

```python
import numpy as np

Furn_test_forecast = pred.predicted_mean
Furn_test_truth = yFurn[datetime(2014,1,1):]
Furn_pred_ci = pred_ci

residual = Furn_test_forecast - Furn_test_truth
mse = (residual ** 2).mean()
print('The Root Mean Squared Error of our forecasts is {}'.format(round(np.sqrt(mse), 4)))
print('The Mean absolute percentage Error is {}'.format(round(np.mean(abs(residual/Furn_test_truth)), 4)))
```

```python
Furn_pred_fc = Furn_res.get_forecast(steps=24)
pred_ci = Furn_pred_fc.conf_int()
ax = yFurn.plot(label='observed', figsize=(14, 7))
Furn_pred_fc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```
